# Problem
Calculate the fourier series for function $f(x) = x^3$

```python
from sympy import *
```

# Part a:
First, we should calculate with 5 sentences.

```python
x = symbols('x')
s = fourier_series(x**3, (x, -2, 2)).truncate(5)
s
```

```python
f = x**3
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(5)
p = plot(f, s1, (x, -2, 2), show=False, legend=False)
p.show()
```

# Part b:
Now we calculate with 11 sentences.

```python
x = symbols('x')
s = fourier_series(x**3, (x, -2, 2)).truncate(11)
s
```

```python
f = x**3
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(11)
p = plot(f, s1, (x, -2, 2), show=False, legend=False)
p.show()
```

# Part c:
Now 21 sentences.

```python
x = symbols('x')
s = fourier_series(x**3, (x, -2, 2)).truncate(21)
s
```

```python
f = x**3
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(21)
p = plot(f, s1, (x, -2, 2), show=False, legend=False)
p.show()
```
