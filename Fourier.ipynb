# Problem
Calculate the fourier series for function $f(x) = x^3$

```python
from sympy import *
from numpy import *
import matplotlib.pyplot as plt
```

# Part a:

```python
x = symbols('x')
s = fourier_series(x**3, (x, -2, 2)).truncate(5)
t = linspace(-2, 2)
plt.plot(t, s)
plt.show()
```
